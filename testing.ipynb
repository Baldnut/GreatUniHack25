### Text Search

```python
url = "https://places.googleapis.com/v1/places:searchText"

complete_query = {
    "textQuery": "York"
}

fields = ["places.displayName", "places.formattedAddress", "places.id", "places.location"]

headers = {
    "Content-Type": "application/json",
    "X-Goog-Api-Key": key,
    "X-Goog-FieldMask": ",".join(fields)
}

response = requests.post(url, json=complete_query, headers=headers)
```

```python
if response.status_code == 200:
    data = response.json()
    for place in data["places"]:
        print(f"Name: {place.get("displayName", {}).get("text", "N/A")}")
        print(f" - Address: {place.get("formattedAddress", "N/A")}")
        print(f" - ID: {place.get("id", "N/A")}")
        print(f" - Location: {place.get("location", "N/A")}")
        print(f"    - latitude: {place.get("location", {}).get("latitude", "N/A")}")
        print(f"    - longitude: {place.get("location", {}).get("longitude", "N/A")}")
else:
    print(f"Error: {response.status_code}")
```

### Nearby search

```python
latitude = 53.961420499999996
longitude = -1.0739108
radius_meters = 2000 # meters
num_results = 10

url = "https://places.googleapis.com/v1/places:searchNearby"
search_types = ["locality"]

complete_query = {
    "includedTypes": search_types,
    "maxResultCount": num_results,
    "locationRestriction": {
        "circle": {
            "center": {
                "latitude": latitude,
                "longitude": longitude
            },
            "radius": radius_meters
        }
    }
}

fields = ["places.displayName", "places.rating"]

headers = {
    "Content-Type": "application/json",
    "X-Goog-Api-Key": key,
    "X-Goog-FieldMask": ",".join(fields)
}

response = requests.post(url, json=complete_query, headers=headers)
```

```python
if response.status_code == 200:
    places = response.json()

    for place in places["places"]:
        print(f"Name: {place.get("displayName", {}).get("text", "N/A")}")
        print(f" - Rating: {place.get("rating", "N/A")}\n")
else:
    print(f"Error: {response.status_code}")
```

### Place Details

```python
place_id = "ChIJ8WWY4UDDeEgR0eRUiomrdEc"

fields = ["displayName", "formattedAddress", "internationalPhoneNumber", "rating"]

url = f"https://places.googleapis.com/v1/places/{place_id}"

headers = {
    "Content-Type": "application/json",
    "X-Goog-Api-Key": key,
    "X-Goog-FieldMask": ",".join(fields)
}

response = requests.get(url, headers=headers)
```

```python
if response.status_code == 200:
    place_details = response.json()
    print("Details for place with ID:", place_id)
    print(f"Name: {place_details.get("displayName", {}).get("text", "N/A")}")
    print(f"  Address: {place_details.get("formattedAddress", "N/A")}")
    print(f"  Phone: {place_details.get("internationalPhoneNumber", "N/A")}")
    print(f"  Rating: {place_details.get("rating", "N/A")}")
else:
    print(f"Error: {response.status_code}")
```
